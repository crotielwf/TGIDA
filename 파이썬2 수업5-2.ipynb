```python
#Table.csv 다운로드 링크 #https://goo.gl/MiH7mR 
#MLB batter.csv 다운로드 링크 #https://goo.gl/G1FnqF

#map() 함수를 다루기 전에
#엑셀파일 /  csv 파일: 컬럼파일이 , 기준으로 나뉘어있음#str.split(',') 사용하면 나눌수있음
#전체 내용을 하나의 큰 리스트로
#각행은 딕셔너리가 되도록
```

```python
from pprint import pprint #pprint는 딕트형 깔끔하게 보여줍니다.
```

```python
with open('table.csv',mode='r') as f:
    #전체 행읽기
    lines=f.readlines()
    
    #첫줄에서 항목별 이름 가져오기
    col_names= lines[0].strip().split(',')#date~ volume이 잡힌다. 
    
    #전체 내용을 담을 빈리스트 정의
    result=[]
    
    #나머지 행에서 전체 항목 데이터 처리
    for line in lines[1:]:
        #해당 행 빈 사전 정의
        row_dict={}
        #해당 행 ''기준 분할
        splited= line.strip().split(',')
        #각 항목마다 해당 행 사전에 key-value 쌍 추가
        for i,col_name in enumerate(col_names):
            row_dict[col_name]=splited[i]
        #해당 행 사전 완성 후 최종 결과 리스트에 추가
        result.append(row_dict)

pprint(result)
```

```python
#이걸 간단히 하는 함수가 map 함수입니다.
#zip() 함수

#map(func,iter) 맵함수는 두개의 전달인자 받습니다.
#iter: (리스트,튜플,...)
```

```python
b=['string','list','dict','tuple']
map_by_endswith=map(lambda x: x.endswith('t'),b)
#람다 함수 생성한거다. 맵함수의 전달인자로 활용
#리스트의 각각 원소마다 't로 끝나는지 여부를 확인하였다.

print(list(map_by_endswith))
```

```python
with open('MLB_batter.csv',mode='r') as f: #역슬래시 넣으면 그대로 이어짐 
    col_names=f.readline().strip().split(',')
    result=list(map(lambda x: dict(zip(col_names,x.strip().split(','))),f.readlines()[1:]))
```

```python
prob_1=[player['Player'] for player in result if int(player['HR']) >=30 and float(player['AVG']) >0.3]

prob_1
```

```python
prob_2=[player['Player'] for player in result if int(player['SO']) >=130 and float(player['H']) <100]

prob_2
```

```python
#여기에 필터함수가 추가된다면?? #filter()
with open('MLB_batter.csv',mode='r') as f:
    col_names=f.readline().strip().split(',')
    result=list(map(lambda x: dict(zip(col_names,x.strip().split(','))),f.readlines()[1:]))
    
prob_1=list(map(lambda x:x['Player'],filter(lambda x:int(x['HR']) >=30 and float(x['AVG']) > 0.3, result)))
```

```python
prob_1
```

```python
prob_2=list(map(lambda x:x['Player'],filter(lambda x:int(x['SO']) >=130 and float(x['H']) < 100, result)))
```

```python
prob_2 #같은결과를 낸다. 필터함수 좋은함수임!
```

```python
#필터예제1
c=['Python','Java','C++','R']
filter_by_startswith=filter(lambda x:x.startswith('C'),c)
print(filter_by_startswith)
print(list(filter_by_startswith))
```

```python
#필터예제2
d=(1,2,3,4)
filter_by_gt_2=filter(lambda x:x>2,d)
print(filter_by_gt_2)
print(list(filter_by_gt_2))
```

```python
#map과 필터의 차이
#map의 결과는 원소마다 전달인자로 함께받은 함수의 적용값을 담은 map 객체
#filter의 결과는 원소마다 전달인자로 함께 받은 함수의 결과가 참인 것을 남긴 filter 객체
```

```python
#map과 필터의 차이예시
e=[['C','Python'],['Java','C++','R'],['Python']]

map_e=map(lambda x:'Python' in x,e)
filter_e=filter(lambda x:'Python' in x,e)
```

```python
print(list(map_e)) #map의 결과는 '전체 원소의 함수 적용 결과'를 모은것 
print(list(filter_e)) #filter의 결과는 기존 원소의 함수 적용 결과가 참인 것'을 모은 것이라는 점을 상기하라
```
